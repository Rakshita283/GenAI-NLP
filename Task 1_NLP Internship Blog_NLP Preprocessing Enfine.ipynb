##__Task 1: Conceptual Understanding__##

1.Difference between "Love" and "love" in NLP --> “Love” (capitalized) might appear at the start of a sentence or as a proper noun (e.g., a brand name, title). --> “love” (lowercase) usually refers to the emotion or verb.

2.What happens if stopwords are not removed?

•Stopwords are common words like the, is, at, and.
•If not removed:
•They increase dimensionality of the dataset unnecessarily.
•They add noise, making models less efficient.
•They can slow down training and reduce accuracy in tasks like topic modeling or keyword extraction.

3.Real-world scenarios where removing stopwords can be harmful

•Sentiment Analysis
•Words like not, never, no are often considered stopwords but are critical for polarity.
•Example: “I do not like this movie” → removing not changes the meaning entirely.
•Question Answering / Search Engines
•Stopwords like who, what, where, when are essential for understanding the type of query.
•Example: “Where is the Eiffel Tower?” → removing where makes the query incomplete.

4.Difference between Stemming and Lemmatization

•Stemming: Mechanical chopping of word endings.
•Lemmatization: Considers context and part of speech to return the correct base form (lemma).

import re
def preprocess_text(text):
    # 1. Convert to lowercase
    text = text.lower()
    # 2. Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)
    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)
    # 4. Handle repeated characters (reduce to max 2)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    # 5. Remove extra spaces and punctuation clutter
    text = re.sub(r'\s+', ' ', text).strip()
    # 6. Tokenize
    tokens = text.split()
    # 7. Remove very short tokens (≤ 2), but keep "no" and "not"
    meaningful = {"no", "not"}
    tokens = [t for t in tokens if len(t) > 2 or t in meaningful]
    # 8. Rejoin tokens
    return " ".join(tokens)

import re
import numpy as np

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)   # remove URLs/emails
    text = re.sub(r'\d+', '', text)                                # remove numbers
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)                     # reduce repeated chars
    text = re.sub(r'\s+', ' ', text).strip()                       # remove extra spaces
    tokens = text.split()
    meaningful = {"no", "not"}
    tokens = [t for t in tokens if len(t) > 2 or t in meaningful]  # remove short tokens
    return tokens, " ".join(tokens)

    # Convert to lowercase
    text = text.lower()

    # Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (reduce to max 2)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    tokens = text.split()

    # Remove very short tokens (≤ 2), but keep "no" and "not"
    meaningful = {"no", "not"}

    return tokens, " ".join(tokens)


def full_pipeline(text_list):
    all_tokens = []
    cleaned_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        cleaned_sentences.append(cleaned)

    return {

    }

import re

def preprocess_text(text):
    # Handle None or empty string
    if not text or text.strip() == "":
        return [], ""

    # Convert to lowercase
    text = text.lower()

    # Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (reduce to max 2)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove emojis and non-alphanumeric symbols (optional safety)
    text = re.sub(r'[^\w\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    tokens = text.split()

    # Remove very short tokens (≤ 2), but keep "no" and "not"
    meaningful = {"no", "not"}
    tokens = [t for t in tokens if len(t) > 2 or t in meaningful]

    # If no tokens remain, return empty safely
    if not tokens:
        return [], ""

    return tokens, " ".join(tokens)
